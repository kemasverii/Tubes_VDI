```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load the uploaded CSV file to examine its structure
file_path = 'dataset_TSMC2014_NYC.csv'
data = pd.read_csv(file_path)

# Display basic information about the dataset and the first few rows
data_info = data.info()
data_head = data.head()

data_info, data_head

# tidak ditemukan data kosong pada dataset NYC
```

```python
#pakai ini

Userid_count = data['userId'].value_counts()
print(Userid_count)
#Tampilkan tempat yang sering dikunjungi oleh 10 user yang paling sering mengunjungi tempat
top_10_user = Userid_count.head(10).index
print(top_10_user)
```

```python
# Identifikasi 10 User ID paling aktif
top_users = data['userId'].value_counts().head(10).index

# Filter data untuk 10 User ID tersebut
top_users_data = data[data['userId'].isin(top_users)]

# Hitung venue (tempat) yang paling sering dikunjungi oleh 10 user
top_venues = top_users_data['venueId'].value_counts().head(10)

# Gabungkan informasi venueId dengan kategori tempat untuk analisis
venue_category_mapping = data.set_index('venueId')['venueCategory'].to_dict()
top_venues_categories = top_venues.index.map(venue_category_mapping)

# Membuat DataFrame untuk visualisasi
top_venues_df = pd.DataFrame({
    'Venue ID': top_venues.index,
    'Venue Category': top_venues_categories,
    'Visit Count': top_venues.values
})

# Membuat warna semakin gelap untuk jumlah kunjungan yang lebih tinggi
colors = sns.color_palette("Blues", len(top_venues_df)) 
# Visualisasi tempat yang sering dikunjungi oleh 10 User ID teraktif
plt.figure(figsize=(12, 6))
plt.bar(top_venues_df['Venue Category'], top_venues_df['Visit Count'], color=colors)
plt.title("Tempat yang Paling Sering Dikunjungi oleh 10 User ID Teraktif", fontsize=16)
plt.ylabel("Jumlah Kunjungan", fontsize=12)
plt.xlabel("Kategori Tempat", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Menampilkan DataFrame hasil
print(top_venues_df)

```

# Visualisasi Jawaban di soal notion untuk pengumpulan terakhir


```python
# Convert utcTimestamp to datetime for easier extraction of components
data['utcTimestamp'] = pd.to_datetime(data['utcTimestamp'], utc=True)

# Extract date, day of the week, and hour from the timestamp
data['date'] = data['utcTimestamp'].dt.date
data['day_of_week'] = data['utcTimestamp'].dt.day_name()
data['hour'] = data['utcTimestamp'].dt.hourz

# Identify the top 10 most visited places (venueId)
top_venues = data['venueCategory'].value_counts().head(10).index

# Filter the data for only the top 10 venues
top_venues_data = data[data['venueCategory'].isin(top_venues)]

# Count visits by venueId and day of the week
visits_by_day = top_venues_data.groupby(['venueCategory', 'day_of_week']).size().unstack(fill_value=0)

# Count visits by venueId and hour
visits_by_hour = top_venues_data.groupby(['venueCategory', 'hour']).size().unstack(fill_value=0)

visits_by_day, visits_by_hour
```

```python
# Create a mapping of venueId to venueCategory for renaming
venue_category_mapping = data.set_index('venueId')['venueCategory'].to_dict()

# Rename indices in the summarized data to use venueCategory instead of venueId
visits_by_day.rename(index=venue_category_mapping, inplace=True)
visits_by_hour.rename(index=venue_category_mapping, inplace=True)

# Display the updated data for confirmation
visits_by_day.head(), visits_by_hour.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the visual style
sns.set(style="whitegrid")

# Plot visits by day of the week for the top 10 venues
plt.figure(figsize=(12, 6))
visits_by_day.T.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='tab10')
plt.title("Kunjungan Berdasarkan Hari di 10 Tempat Teramai", fontsize=16)
plt.ylabel("Jumlah Kunjungan", fontsize=12)
plt.xlabel("Hari dalam Seminggu", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Venue ID", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Heatmap for visits by hour
plt.figure(figsize=(14, 7))
sns.heatmap(visits_by_hour, cmap="YlGnBu", linewidths=.5, annot=False, cbar_kws={'label': 'Jumlah Kunjungan'})
plt.title("Pola Kunjungan Berdasarkan Jam di 10 Tempat Teramai", fontsize=16)
plt.ylabel("Venue ID", fontsize=12)
plt.xlabel("Jam (24-hour)", fontsize=12)
plt.tight_layout()
plt.show()

```

# Tempat teramai dikunjungi

```python
import matplotlib.pyplot as plt

# Distribusi kategori tempat
category_counts = data['venueCategory'].value_counts().head(10)  # Top 10 categories

# Visualisasi
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='skyblue')
plt.title('Top 10 Venue Categories', fontsize=16)
plt.xlabel('Venue Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
for i, count in enumerate(category_counts):
    plt.text(i, count + 1, str(count), ha='center', va='bottom')
plt.show()

category_counts
#tambahkan total count pada masing-masing bar
```

Bagian ini menjelaskan persebaran tempat-tempat yang ramai dikunjungi,terlihat pada bar chart  tempat seperti bar serta rumah dan kantor menjadi 3 tempat teratas yang ramai dikunjungi. Total pengujung bar berjumlah 15978 pengunjung,Home (private) sebanyak 15382 dan Office sebanyak 12740

```python
# instal dulu library folium -> pip install folium
import folium

# Membuat peta lokasi populer
map_center = [data['latitude'].mean(), data['longitude'].mean()]
popularity_map = folium.Map(location=map_center, zoom_start=12)

# Menambahkan titik lokasi (sampling untuk efisiensi)
sampled_data = data.sample(1000)  # Mengambil sampel 1000 baris untuk peta
for _, row in sampled_data.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=2,
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
    ).add_to(popularity_map)

# Simpan peta ke file untuk ditampilkan
map_file_path = 'popularity_map.html'
popularity_map.save(map_file_path)

map_file_path
```
